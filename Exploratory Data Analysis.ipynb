## merging data 

```python
table1 = pd.DataFrame({"P_ID" : (1,2,3,4,5,6,7,8),
                     "gender" : ("male", "male", "female","female",
                                "female", "male", "female", "male"),
                     "height" : (71,73,64,64,66,69,62,72),
                     "weight" : (175,225,130,125,165,160,115,250)})
```

```python
table1
```

```python
table2 = pd.DataFrame({"P_ID" : (1, 2, 4, 5, 7, 8, 9, 10),
                     "sex" : ("male", "male", "female","female",
                            "female", "male", "male", "female"),
                     "visits" : (1,2,4,12,2,2,1,1),
                     "checkup" : (1,1,1,1,1,1,0,0),
                     "follow_up" : (0,0,1,2,0,0,0,0),
                     "illness" : (0,0,2,7,1,1,0,0),
                     "surgery" : (0,0,0,2,0,0,0,0),
                     "ER" : ( 0,1,0,0,0,0,1,1) } ) 

table2
```

```python
combined1=pd.merge(table1,table2,how="inner",on="P_ID")
combined1
```

```python
combined2=pd.merge(table1,table2,how="outer",on="P_ID")
```

```python
combined2
```

```python
table1.rename(columns={"gender":"sex"},inplace=True)
```

```python
table1
```

```python
combined3=pd.merge(table1,table2,how="inner",on=["P_ID","sex"])
```

```python
combined3
```

## Frequency tables 

```python
titanic=pd.read_csv("./data/titanic/train.csv")
```

```python
titanic
```

```python
survived_class=pd.crosstab(index=titanic.Survived,
                          columns=titanic.Pclass,margins=True)
survived_class

survived_class.index=["died","survived","col_total"]
survived_class.columns=["class1","class2","class3","row_total"]
survived_class
```

```python
survived_class/survived_class.loc["col_total","row_total"]
```

```python
survived_class/survived_class.loc["col_total"]
```

```python
survived_class.div(survived_class["row_total"],axis=0)
```

```python
survived_class
```

```python
sex_class_surv=pd.crosstab(index=titanic.Survived,
                          columns=[titanic.Pclass,titanic.Sex],
                          margins=True)
```

```python
sex_class_surv
```

```python
sex_class_surv/sex_class_surv.loc["All"]
```

```python
sex_class_surv.div(sex_class_surv["All"],axis=0)
```
